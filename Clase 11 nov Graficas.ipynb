### Ver si existe una correlaci√≥n entre dos variables 'wingspan' vs 'CWDistance'

```python
df = pd.read_csv('datasets/Cartwheeldata.csv')
df.head()
```

```python
pip install seaborn
```

```python
import seaborn as sns
```

```python
sns.lmplot(x = 'Wingspan', 
           y = 'CWDistance', 
           data = df,
           fit_reg = False, # parametro para eliminar la linea de regresion
           hue = 'Gender' # parametro para distinguir los distintos tipos de la categor√≠a
          )
plt.title('Scatter plot of Wingspan vs CWDistance', fontsize = 15)
```

```python
sns.swarmplot(x = 'Gender', 
              y = 'CWDistance', 
              data = df
             )
plt.title('Swarm plot of Gender vs CWDistance', fontsize = 16)
```

## Estadistica descriptiva

```python
df.describe()
df.columns
```

```python
sns.boxplot(data = df.loc[:,['Age', 'Gender', 'GenderGroup', 'Glasses', 'GlassesGroup',
       'Height', 'Wingspan', 'CWDistance']]
           )
plt.xticks(fontsize = 17, rotation = 45)
plt.yticks(fontsize = 17)
plt.title('Boxplots', fontsize = 17)
```

```python
sns.boxplot(data = df.loc[df['Gender']=='F',
                          ['Age', 'Gender', 'GenderGroup', 'Glasses', 'GlassesGroup',
       'Height', 'Wingspan', 'CWDistance']]
           )
plt.xticks(fontsize = 17, rotation = 45)
plt.yticks(fontsize = 17)
plt.title('Boxplots', fontsize = 17)
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(1,2,1) # 1 rengl√≥n, 2 columnas, 1st position
sns.boxplot(data = df.loc[df['Gender'] == 'F',['Wingspan']]
           )

ax2 = fig.add_subplot(1,2,2) # 1 rengl√≥n, 2 columnas, 2nd position
sns.boxplot(data = df.loc[df['Gender'] == 'M',['Wingspan']]
           )
ax1.title.set_text('Genero F Wingspan')
ax2.title.set_text('Genero M Wingspan')
```
