```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
%ls datasets/
```

```python
df = pd.read_csv('datasets/tips.csv')
df.head()
```

## Pregunta de investigación 1 - https://www.itl.nist.gov/div898/handbook/eda/section3/eda366.htm
# Qué tipo de distribuciones tienen los datos

```python
#df.columns
sns.distplot(df['total_bill'],
             #kde = False, #Whether to plot a gaussian kernel density estimate.
             #bins = 5 #cuantas barras del histograma. Default = 10
            ).set_title('Histograma de total bill', fontsize = 16)
plt.show()
```

# Conclusion: 
### La variable 'total_bill' parece 'Weibull Distribution'

```python
sns.boxplot(x = df['total_bill'],
            y = df['day'],
           whis = (0,100)
           ).set_title('Box plot of total_bill')
```
