# Scatter Plots

```python
sns.lmplot(x = 'peso', 
           y = 'talla', 
           data = df,
           fit_reg = False, # parametro para eliminar la linea de regresion
           col = 'sexo' # parametro para distinguir los distintos tipos de la categor√≠a
          )
plt.title('Scatter plot of Wingspan vs CWDistance', fontsize = 15)
```

# Histograma
